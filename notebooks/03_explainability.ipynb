# Explainability

## Overview

Our goal was to predict corn yield per acre using agronomic and management-related features. We used an XGBoost regression model because it performs well on structured tabular data and can capture nonlinear relationships between variables such as planting timing, nitrogen application, and irrigation.

---

## Data Preprocessing and Cleaning

The original dataset contained missing values, sparse columns, and several features that could lead to misleading model behavior. We performed the following cleaning steps:

- Removed rows with missing target values (`yieldPerAcre`)
- Dropped weak or sparse columns such as `qrCode`, `plotNumber`, and others with limited usefulness
- Removed structural columns such as `index`, `row`, `range`, `block`, and `plotLength` to avoid encoding field layout instead of real agronomic signal
- Eliminated `experiment` and `location` features to prevent shortcut learning and improve generalization

These steps ensured that the model focused on meaningful inputs rather than dataset artifacts.

---

## Feature Engineering

To improve model performance and interpretability, we created several engineered features based on domain intuition:

### Planting Date Features

We transformed `plantingDate` into:

- `planting_month`
- `planting_dayofyear`
- `is_early_season` (binary indicator for early planting)

This allowed the model to capture seasonal effects more effectively rather than treating each date as a unique category.

---

### Nitrogen Features

We engineered features to capture more realistic fertilizer behavior:

- `nitrogen_squared`: models nonlinear response to nitrogen (e.g., diminishing returns)
- `nitrogen_x_irrigation`: interaction between nitrogen and irrigation

These features help represent real-world agricultural dynamics, where fertilizer effectiveness often depends on water availability.

---

## Feature Importance Analysis

Feature importance from the model shows that the most influential variables include:

- Planting day of year
- Nitrogen application (linear and nonlinear)
- Irrigation
- Nitrogen–irrigation interaction
- Planting month

This suggests that yield is strongly influenced by **when crops are planted**, **how nutrients are applied**, and **water availability**.

---

## Preventing Leakage and Improving Generalization

During development, we observed that location-based features were dominating the model. This indicated that the model was relying on geographic shortcuts rather than learning general agronomic patterns.

To address this:

- We removed `location` entirely
- We removed `experiment` identifiers
- We dropped structural grid columns (`row`, `range`, `block`, etc.)

This forced the model to learn relationships that generalize across environments rather than memorizing specific fields.

---

## Model Behavior and Limitations

While the model performs reasonably well, cross-validation results show variability across different splits. This indicates:

- Yield prediction is inherently noisy and complex
- Some environmental or genetic factors may not be fully captured
- The model should be interpreted as a **decision-support tool**, not an exact predictor

---

## Key Insights

From the model and feature analysis, we observe:

- Earlier planting tends to improve predicted yield
- Nitrogen has a nonlinear effect on yield
- Irrigation enhances the effectiveness of nitrogen
- Yield is driven more by management decisions than by location in the final model

---

## Conclusion

By removing shortcut features and engineering meaningful agronomic variables, we created a model that is both interpretable and aligned with real-world farming practices. The final pipeline prioritizes generalizable patterns such as planting timing, nutrient management, and irrigation, making it more robust and useful for decision-making.
